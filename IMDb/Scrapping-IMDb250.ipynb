```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bts
import requests
import csv
```

```python
def getAndParseURL(url):        
    result = requests.get(url,headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36'})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
PRODUCT_URL = []
def products():
    html = getAndParseURL('https://www.imdb.com/chart/top/')
    for b in html.findAll("td",{"class":"titleColumn"}): #column
        PRODUCT_URL.append('https://www.imdb.com/'+b.a["href"])
    return PRODUCT_URL
products()
```

```python
ID=[]
Name=[]
Gender=[]
Puan=[]
Director=[]
Realase_Date=[]
Reward=[]
Sum=[]

```

```python
for i in PRODUCT_URL[::]:
    ID=i.split("-P")[-1]
    html = getAndParseURL(i)
    
    name=html.find("h1",attrs={"data-testid":"hero-title-block__title"}).text
    Name.append(name)
    
    gender=html.find("span",attrs={"class":"ipc-chip__text"}).text
    Gender.append(gender)
    
    puan=html.find("span",attrs={"class":"AggregateRatingButton__RatingScore-sc-1ll29m0-1 iTLWoV"}).text
    Puan.append(puan)
    
    summary=html.find("span",attrs={"class":"GenresAndPlot__TextContainerBreakpointXL-cum89p-2 gCtawA"}).text
    Sum.append(summary)
    
    director=html.find("a",attrs={"class":"ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link"}).text
    Director.append(director)
    
    rd=html.find("a",attrs={"class":"ipc-link ipc-link--baseAlt ipc-link--inherit-color TitleBlockMetaData__StyledTextLink-sc-12ein40-1 rgaOW"}).text
    Realase_Date.append(rd)
    
    reward=html.find("span",attrs={"class":"ipc-metadata-list-item__list-content-item"}).text
    Reward.append(reward)
    
```

```python
data= pd.DataFrame({"Name":Name,"Gender":Gender,"Star":Puan,"Summary":Sum,"Director":Director,"Realase Date":Realase_Date,"Reward":Reward})
data
```

```python
data.to_csv('MovieTop250.csv', index=False, encoding='utf-8')
```
