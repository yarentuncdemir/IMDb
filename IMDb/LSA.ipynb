```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
moviedata=pd.read_csv('MovieTop250.csv')
```

```python
sum_data = pd.DataFrame({
    'sum' : moviedata['Summary'],
})

stop_words = set(stopwords.words('english'))
      
def clean_text(text):
    text = re.sub("[^a-zA-Z]", " ", text)
    text = ' '.join(text.split(','))
    text = text.lower()
    no_stopword_text = [w for w in text.split() if not w in stop_words]
    return ' '.join(no_stopword_text)
    

sum_d = sum_data['sum'].apply(clean_text)
sum_d
```

```python
numpy_array = sum_d.to_numpy()
np.savetxt("test_file.txt", numpy_array, fmt = "%s")
```

```python
bodytext=open("test_file.txt",'r')
body=bodytext.readlines()
```

```python
body
```

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
vectorizer=CountVectorizer()
bag_of_words=vectorizer.fit_transform(body)
bag_of_words.todense()
```

```python
svd=TruncatedSVD(n_components=2)
lsa=svd.fit_transform(bag_of_words)
```

```python
topic=pd.DataFrame(lsa,columns=["topic1","topic2"])
topic["body"]= body
display(topic[["body","topic1","topic2"]])
```

```python
dictionary = vectorizer.get_feature_names()
```

```python
encoding_matrix=pd.DataFrame(svd.components_,["topic1","topic2"],columns=dictionary).T
```

```python
encoding_matrix['topic3']=np.abs(encoding_matrix['topic1'])
encoding_matrix['topic4']=np.abs(encoding_matrix['topic2'])
encoding_matrix.sort_values('topic3',ascending=False)
```

```python
final_matrix=encoding_matrix.sort_values('topic3',ascending=False)
final_matrix[["topic3","topic4"]]
```
