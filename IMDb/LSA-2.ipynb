```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
moviedata=pd.read_csv('MovieTop250.csv')
```

```python
name_data = pd.DataFrame({
    'name' : moviedata['Name'],
})
```

```python
stop_words = set(stopwords.words('english'))
      
def clean_text(text):
    text = re.sub("[^a-zA-Z]", " ", text)
    text = ' '.join(text.split(','))
    text = text.lower()
    no_stopword_text = [w for w in text.split() if not w in stop_words]
    return ' '.join(no_stopword_text)
    

name_d = name_data['name'].apply(clean_text)
name_d
```

```python
numpy_array = name_d.to_numpy()
np.savetxt("name_file.txt", numpy_array, fmt = "%s")
```

```python
bodytext=open("name_file.txt",'r')
body=bodytext.readlines()
```

```python
body
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
vectorizer = TfidfVectorizer(use_idf=True, ngram_range=(1, 3))
X = vectorizer.fit_transform(body)
```

```python
X.shape

lsa = TruncatedSVD(n_components=10, n_iter=10)
lsa.fit(X)
lsa.components_[0]

terms = vectorizer.get_feature_names()
for i, comp in enumerate(lsa.components_): 
    termsInComp = zip (terms,comp)
    sortedTerms =  sorted(termsInComp, key=lambda x: x[1], reverse=True) [:5]
    print("Concept %d:" % i)
    for term in sortedTerms:
        print(term[0])
    print(" ")
```
